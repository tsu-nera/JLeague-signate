# pythagorean expectation

```python
import numpy as np
import pandas as pd
import re
```

```python
###データ読込(結合したデータ)
train = pd.read_csv("~/repo/study/JLeague/data/train_new3.csv")
test = pd.read_csv("~/repo/study/JLeague/data/test_new3.csv")
test_id = pd.read_csv("~/repo/study/JLeague/motodata/test.csv")["id"]
test_id_all = test["id"]
```

```python
for i, m in enumerate(train["match"]):
    match = re.findall("第(\w+)節第(\w+)日", str(m))[0]
    train.loc[i, "match"] = int(match[0])
for i, m in enumerate(test["match"]):
    match = re.findall("第(\w+)節第(\w+)日", str(m))[0]
    test.loc[i, "match"] = int(match[0])
```

```python
train = train[["id","year","match", "home", "away", "home_score", "away_score"]]
test = test[["year", "match", "home", "away", "home_score", "away_score"]]
```

```python
teams = train.home.unique()
```

```python
def pythagorean_expectation(runs_scored, runs_allowed):
    return runs_scored**2/((runs_scored**2)+(runs_allowed)**2)
```

```python
runs_scored_tbl = pd.DataFrame(columns=teams)
```

```python
for i in train.iterrows():
    year = i[1]["year"]
    match = i[1]["match"]
    index = match + (year-2012)*42
    home = i[1]["home"]
    home_score = i[1]["home_score"]
    away = i[1]["away"]
    away_score = i[1]["away_score"]
    if match == 1:
        runs_scored_tbl.loc[index, home] = 0
        runs_scored_tbl.loc[index, away] = 0
    else:
        runs_scored_tbl.loc[index, home] = home_score + runs_scored_tbl.loc[index-1, home]
        runs_scored_tbl.loc[index, away] = away_score + runs_scored_tbl.loc[index-1, away]
```

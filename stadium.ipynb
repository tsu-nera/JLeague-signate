```python
import pandas as pd
import geocoder
```

```python
stadium = pd.read_csv("../JLeague/data/stadium.csv")
```

```python
for i, addr in enumerate(stadium["address"]):
    g = geocoder.google(addr)
    stadium.loc[i, "latitude"] = g.latlng[0]
    stadium.loc[i, "longitude"] = g.latlng[1]
```

```python
# https://uub.jp/prf/prefbase.html
prefecture = pd.read_csv("../JLeague/data/area.csv")
```

```python
prefecture
```

```python
stadium2 = pd.merge(stadium, prefecture, how="left", on="prefecture")
```

```python
stadium2
```

```python
city = pd.read_csv("../JLeague/data/city.csv")
```

```python
for i, addr in enumerate(city["city"]):
    g = geocoder.google(addr)
    city.loc[i, "c_latitude"] = g.latlng[0]
    city.loc[i, "c_longitude"] = g.latlng[1]
```

```python
city
```

```python
city.to_csv("../JLeague/data/city.csv", index=False, header=True, float_format='%g')
```

```python
stadium3 = pd.merge(stadium2, city, how="left", on="prefecture")
```

```python
import math
def deg2rad(deg):
    return( deg * (2 * math.pi) / 360 )

def Hubeny(lat1, lon1, lat2, lon2) :
    a =6378137.000
    b =6356752.314140
    e =math.sqrt( (a**2 - b**2) / a**2 )
    e2 =e**2
    mnum =a * (1 - e2)

    my =deg2rad((lat1+lat2) /2.0)
    dy =deg2rad( lat1-lat2)
    dx =deg2rad( lon1-lon2)
    sin =math.sin(my)
    w =math.sqrt(1.0-e2 * sin *sin)
    m =mnum /(w *w *w)
    n =a/w
    dym =dy*m
    dxncos=dx*n*math.cos(my)
    return( math.sqrt( dym**2 + dxncos**2) )
```

```python
for i, item in enumerate(stadium3.iterrows()):
    lat1 = item[1]["latitude"]
    lon1 = item[1]["longitude"]
    lat2 = item[1]["c_latitude"]
    lon2 = item[1]["c_longitude"]
    stadium3.loc[i, "distance"] = Hubeny(lat1, lon1, lat2, lon2) 
```

```python
stadium3.to_csv("../JLeague/data/stadium3.csv", index=False, header=True, float_format='%g')
```
